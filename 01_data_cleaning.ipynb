```python
import pandas as pd
import numpy as np
import os

# 1. Definisco il percorso del file Excel completo
file_path = os.path.join("data", "online_retail_II.xlsx")
print("Caricamento del dataset in corso... ⏳")

# 2. Carico entrambi i fogli impostando i codici come stringhe (str) per non perdere gli zeri e non avere il .0
df_2009_2010 = pd.read_excel(file_path, sheet_name="Year 2009-2010", dtype={'Customer ID': str, 'StockCode': str})
df_2010_2011 = pd.read_excel(file_path, sheet_name="Year 2010-2011", dtype={'Customer ID': str, 'StockCode': str})

# 3. Unisco i due fogli del biennio
df = pd.concat([df_2009_2010, df_2010_2011], ignore_index=True)

# 4. Assicuro che la colonna delle date sia riconosciuta correttamente come data di Pandas
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# 5. Conferma di successo e dimensioni
print("Dataset importato correttamente e tipizzato secondo la checklist! 🎉")
print(f"Righe totali: {df.shape[0]} | Colonne totali: {df.shape[1]}")

# 6. Mostro l'anteprima per verificare che il .0 su Customer ID sia sparito
df.head()
```

```python
# Task 1.2: Informazioni generali e statistiche descrittive
print("📋 --- INFO STRUTTURALI ---")
df.info()

print("\n📊 --- STATISTICHE DESCRITTIVE ---")
print(df.describe())
```

```python
# FASE 1 - Task 1.3: Conteggio nulli e percentuale su Customer ID
print("❓ --- VALORI NULLI PER OGNI COLONNA ---")
valori_nulli = df.isnull().sum()
print(valori_nulli)

print("\n📉 --- PERCENTUALE NULLI CUSTOMER ID ---")
pct_null_customer = (valori_nulli['Customer ID'] / len(df)) * 100
print(f"Percentuale di Customer ID nulli: {pct_null_customer:.2f}%")
```

```python
# Task 1.4: Controllo dei duplicati e rimozione
print("👥 --- RIGHE DUPLICATE RILEVATE ---")
duplicati_totali = df.duplicated().sum()
print(f"Numero di righe completamente duplicate: {duplicati_totali}")

# Rimuovo i duplicati salvando il risultato dentro 'df'
df = df.drop_duplicates()

print("\n✅ --- VERIFICA POST-PULIZIA ---")
duplicati_rimasti = df.duplicated().sum()
print(f"Righe duplicate rimaste: {duplicati_rimasti}")
print(f"Nuovo numero totale di righe nel dataset: {len(df)}")
```
